Clusteren van hoog-dimensionale data
====================================

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

Mijn scriptie gaat over het clusteren van hoog-dimensionale data. Allereerst leg ik clusteren uit en dan waarom het lastig is om hoog-dimensionale data te clusteren. 

Clusteren
---------
Clusteren betekent dat je de data in groepen probeert te verdelen en wordt vaak 'unsupervised' gedaan. Unsupervised betekent dat je niet weet wat de 'ware' verdeling in groepen is. Als je gegevens hebt van mensen en hun inkomen dan zou een logische uitkomst van een clustering bijvoorbeeld zijn dat je een groep mannen en vrouwen vindt. Maar het zou ook zo kunnen zijn dat het cluster programma een onderverdeling naar opleiding vind. Van te voren weet je bij 'unsupervised' clustering dus niet wat de groepen zijn. Clusteren wordt daarom vaak gebruikt om de data te verkennen en te kijken of het kan worden samengevat. 

Over het algemeen wordt clusteren gedefinieerd als het verdelen van de data in groepen met veel gelijkheid binnen de groepen, en veel verschillen tussen de groepen. In het onderstaande figuur is data te zien dat afkomstig is van twee groepen die aardig gemakkelijk te herkennen zijn. 

```python
mu_men = np.array([25., 20000.])
sigma_men = np.array([[7**2, 25**2], [25**2, 2000**2]])
mu_women = np.array([45., 18000,])
sigma_women = np.array([[5**2, 6**2], [6**2, 800**2]])
men = np.random.multivariate_normal(mu_men, sigma_men, 100)
women = np.random.multivariate_normal(mu_women, sigma_women, 100)
data = pd.DataFrame(np.vstack((men, women)), range(200), ["Leeftijd", "Inkomen"])
```

De 'ware' clustering is echter moeilijk te achterhalen. Een deel van de groepen overlappen en daardoor is achteraf niet meer goed te zeggen wat bij wie hoort. Als de gemiddelden van de groep bekend zijn is het volgende bijvoorbeeld een logische clustering.

```python
plt.title("Data")
plt.scatter(data["Leeftijd"], data["Inkomen"])
plt.xlabel("Leeftijd")
plt.ylabel("Inkomen")
```

```python
dist_men = np.linalg.norm(data - mu_men, 2, 1)
dist_women = np.linalg.norm(data - mu_women, 2, 1)
is_men = dist_men < dist_women
is_women = dist_women <= dist_men
```

```python
plt.title("Gevonden clustering")
plt.scatter(data[is_men]["Leeftijd"], data[is_men]["Inkomen"], c='b')
plt.scatter(data[is_women]["Leeftijd"], data[is_women]["Inkomen"], c='r')
plt.xlabel("Leeftijd")
a = plt.ylabel("Inkomen")
```

En dan blijkt dat er iets van de 'ware' clustering verloren is gegaan omdat de groepen overlappen.

```python
plt.title("Ware clustering")
plt.scatter(men[:, 0], men[:, 1] , c='b')
plt.scatter(women[:, 0], women[:, 1], c='r')
plt.xlabel("Leeftijd")
a = plt.ylabel("Inkomen")
```

Waarom is het clusteren van hoog-dimensionale data lastig
---------------------------------------------------------

Om dit te begrijpen moeten we eerst kijken naar wat er vaak gebeurt om laag-dimensionale data te clusteren. Dit is data die maar weinig features/variablen heeft. Zojuist hebben we gekeken naar data met 2 variabelen (de x en y-as), dus data met dimensie 2. Een deel van die data ziet er in de tabel als volgt uit.

```python
data[:10]
```

We kunnen meten hoe groot het verschil is tussen alle personen door met een lineaal de afstand tussen de punten in de plot te bekijken. Dit wordt ook wel de afstand tussen twee datapunten genoemd. Als we bijvoorbeeld alleen naar de verschillen tussen de leeftijden (dus alleen op de horizontale as) kijken dan levert dit $200 * 200 * 1/2$ aan verschillen op, namelijk tussen elk paar van leeftijden. De waarden daarvan kunnen we laten zien in een histogram.

```python
distances = pdist(data["Leeftijd"].reshape((-1, 1)))
plt.hist(distances)
plt.xlabel("Verschil in leeftijd")
a = plt.ylabel("Aantal")
```

Om nu rekening te houden met de echte afstand tussen punten moeten ook rekening houden met het inkomen. Probleem is dat deze zich op een heel andere schaal bevind. Dit wordt vaak opgelost door de data te normaliseren. Het gemiddelde wordt van de punten afgetrokken en het restant wordt gedeeld door de standaard afwijking. Daardoor is het nieuwe gemiddelde $0$ en de nieuwe standaard afwijking $1$. Dat levert de volgende data op.

```python
data_norm = (data - data.mean(0)) / data.std(0)
plt.scatter(data_norm["Leeftijd"], data_norm["Inkomen"])
plt.title("Genormaliseerde data")
plt.xlabel("(Leeftijd - gemiddelde leeftijd) / std leeftijd")
a = plt.ylabel("(Inkomen - gemiddeld inkomen) / std inkomen)")
```

Nu hebben beide features dezelfde schaal en kunnen we opnieuw de afstanden met alleen de leeftijden berekenen en daarna het verschil zien als we ook rekening houden met de inkomsten. 

```python
distances_age_norm = pdist(data_norm["Leeftijd"].reshape((-1, 1)))
distances_both_norm = pdist(data_norm[["Leeftijd", "Inkomen"]])
a = plt.hist(distances_age_norm)
plt.xlim([0, 10])
plt.title("Genormaliseerde afstand tussen leeftijden")
plt.xlabel("Afstand")
plt.ylabel("Aantal")
plt.figure()
a = plt.hist(distances_both_norm)
plt.xlim([0, 10])
plt.title("Genormaliseerde afstand tussen leeftijden en inkomsten")
plt.xlabel("Afstand")
a = plt.ylabel("Aantal")
```

Waar bij alleen de leeftijden nog de meeste afstanden tussen de $0$ en $0.5$ liggen is dat bij de afstand tussen zowel de leeftijd en inkomsten niet meer zo. Een manier om dit te snappen is te bedenken dat minder punten zowel de zelfde leeftijd als inkomsten hebben dan een van tweeen. Dit process stop niet bij twee dimensies. We kunnen nog een derde toevoegen, bijvoorbeeld het aantal kinderen. 

```python
data["Kinderen"] = np.hstack([np.random.randint(0, 3, 100), np.random.randint(2, 5, 100)])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
a = ax.scatter(data["Leeftijd"], data["Inkomen"], data["Kinderen"])
```

Op de hoogte as is nu het aantal kinderen te zien, dat varieert van $0$ tot $5$. Mannen hebben tussen de $0$ en $3$ kinderen en vrouwen hebben tussen de $2$ en $5$ kinderen. Hoewel dit een onrealistische beeld is, geeft ons dit meer informatie om de groepen te scheiden. Laten we kijken wat het met de afstanden doet die we nu meten als afstand tussen twee 3d punten in de genormaliseerde data. 

```python
data_norm["Kinderen"] = (data["Kinderen"] - data["Kinderen"].mean(0)) / data["Kinderen"].std(0)
distance_three_norm = pdist(data_norm[["Leeftijd", "Inkomen", "Kinderen"]])
a = plt.hist(distance_three_norm)
plt.xlim([0, 10])
plt.title("Genormalizeerde afstand tussen mensen")
plt.xlabel("Afstand")
a = plt.ylabel("Aantal")
```

Duidelijk is te zien dat nu vrijwel niets meer echt dichtbij is. Laten we kijken als we 5 willekeurige features toevoegen. Nu wordt het moeilijk te visualiseren omdat we dan in totaal 13 dimensies hebben maar het uitrekenen van de afstand gaat op de zelfde manier. 

```python
for i in range(0, 10):
    data["Random"+str(i)] = data["Leeftijd"] * np.random.random(1) + np.random.random(200) * 30
data_norm = (data - data.mean()) / data.std()
distance_all_norm = pdist(data_norm)
a = plt.hist(distance_all_norm)
plt.xlim([0, 10])
plt.title("Genormalizeerde afstand tussen mensen")
plt.xlabel("Afstand")
a = plt.ylabel("Aantal")
```

Nu zijn er helemaal geen mensen meer die een kleine afstand to elkaar hebben. Dat komt omdat ze altijd wel op een bepaalde feature van elkaar verschillen, het komt maar heel weinig voor dat mensen bij alle features een gelijke score hebben.

Hoe meer features er worden toegevoegd hoe minder punten dicht bij elkaar liggen. Het verschil tussen punten die ver weg liggen en punten dicht bij liggen zegt steeds minder. Als er teveel features worden toegevoegd kun je niet meer echt spreken van dichtbij of ver weg, alle punten zijn gewoon verschillend. Dit fenomeen word de curse-of-dimensionality genoemd. Door extra features, en dus informatie, toe te voegen zou het makkelijker moeten zijn om punten te clusteren maar het wordt juist moeilijker doordat de afstandmaat minder zegt. 

Sparse subspace clustering
==========================

Sparse subspace clustering is een manier om toch om te gaan met data met een hoge dimensie. Anders dan traditionele methodes maakt het geen gebruik van de afstand tussen punten om ze te clusteren. Het maakt wel gebruik van de zogenaamde self-expresiveness eigenschap. Met deze eigenschap is het mogelijk om punten in dezelfde subspace aan elkaar te relateren, hoewel de dimensie erg hoog kan zijn. Deze informatie wordt vervolgens gebruikt om de data te clusteren op de normale manier.
